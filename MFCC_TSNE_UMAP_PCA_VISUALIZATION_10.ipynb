#Importing and loading MFCC dataset

```python
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mfcc_data=pd.read_csv('/content/mfcc_features_Indian_10.csv')
mfcc_data
```

```python
# Extract the features (all columns except the last one which is 'label')
features = mfcc_data.iloc[:, :-1].values  # All rows, all columns except the last
labels = mfcc_data['label'].values        # All rows, 'label' column
```

#PCA for Dimensionality Reduction and visualizing in 2d

```python
from sklearn.decomposition import PCA
```

```python
# Perform PCA to reduce to 2 dimensions
pca = PCA(n_components=10,random_state=42)
features_pca_10 = pca.fit_transform(features[:30])
```

```python
# Get the first 10 eigenvalues
eigenvalues = pca.explained_variance_

# Display the first 10 eigenvalues
print("First 10 eigenvalues of MFCC PCA method:")
print(eigenvalues)
```

```python
# Perform PCA to reduce to 2 dimensions
pca = PCA(n_components=2,random_state=42)
features_pca = pca.fit_transform(features)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(features_pca)
```

```python
# Define labels in the desired order
desired_label_order = ['Flute', 'Nadaswaram','Shehnai' ,'Piano','Thavil','Mridangam', 'Santoor', 'Veena','Guitar','Violin']

# Define a color map for each label
color_map = {
    'Flute': 'darkblue',
    'Nadaswaram': 'yellow',
    'Shehnai':'pink',
    'Thavil': 'red',
    'Piano': 'black',
    'Mridangam':'orange',
    'Santoor': 'brown',
    'Veena': 'green',
    'Guitar':'purple',
    'Violin':'violet'
}

# Create a scatter plot with specified colors for labels
plt.figure(figsize=(10, 8))
for label in desired_label_order:
    label_indices = np.array(labels) == label  # Find indices corresponding to the label
    plt.scatter(
        scaled_data[label_indices, 0],  # X-axis (PCA Component 1)
        scaled_data[label_indices, 1],  # Y-axis (PCA Component 2)
        c=color_map[label],  # Use the specified color for the label
        label=label, alpha=0.9
    )

# Add titles and legend
plt.title("PCA: 2D Visualization of MFCC Features", fontsize=16)
plt.xlabel("Principal Component 1", fontsize=12)
plt.ylabel("Principal Component 2", fontsize=12)
plt.legend(title="Labels", loc='best')
plt.grid()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Assuming `features_pca` contains PCA-transformed data and `labels` contains cluster labels
score = silhouette_score(scaled_data, labels)
print(f"Silhouette Score: {score}")
```

```python
from scipy.spatial.distance import pdist, squareform
import pandas as pd

# Assuming `features_pca` is a NumPy array of shape (n_samples, n_features)
# Step 1: Compute pairwise distances
pairwise_distances = squareform(pdist(scaled_data, metric='euclidean'))

# Step 2: Convert to a DataFrame for better readability
df_pairwise = pd.DataFrame(pairwise_distances)

# Optional: Add labels or indices if desired
df_pairwise.index = labels  # or use a custom index
df_pairwise.columns = labels  # same here

# Step 3: Round for readability
df_pairwise = df_pairwise.round(2)

# Print or view the DataFrame
df_pairwise
```

#TSNE Plot for the Musical Instruments

#UMAP Visualization of MFCC Features

```python
import umap.umap_ as umap
```

```python
# Initialize UMAP with 2 components for a 2D plot
umap_reducer = umap.UMAP(n_components=2, n_neighbors=30, min_dist=0.25)


# Fit and transform the features using UMAP
umap_results = umap_reducer.fit_transform(features)
```

```python
scaler = MinMaxScaler()
scaled_data_umap = scaler.fit_transform(umap_results)
```

```python
# Create a scatter plot with specified colors for labels
plt.figure(figsize=(10, 8))
for label in desired_label_order:
    label_indices = np.array(labels) == label  # Find indices corresponding to the label
    plt.scatter(
        scaled_data_umap[label_indices, 0],  # X-axis (UMAP Component 1)
        scaled_data_umap[label_indices, 1],  # Y-axis (UMAP Component 2)
        c=color_map[label],  # Use the specified color for the label
        label=label, alpha=0.9
    )

# Add titles and legend
plt.title("UMAP: 2D Visualization of MFCC Features", fontsize=16)
plt.xlabel("Dimension 1", fontsize=12)
plt.ylabel("Dimension 2", fontsize=12)
plt.legend(title="Labels", loc='best')
plt.grid()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Assuming `umap_results` contains UMap-transformed data and `labels` contains cluster labels
score = silhouette_score(scaled_data_umap, labels)
print(f"Silhouette Score: {score}")
```
